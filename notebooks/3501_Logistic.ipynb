### Entrenamiento

```python
from sklearn.model_selection import train_test_split

target_column = 'CO2_emissions'  
classification_target = 'fuel_type' 

X_reg = df.drop(columns=[target_column])
y_reg = df[target_column]

X_clf = df.drop(columns=[classification_target])
y_clf = df[classification_target]

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(X_clf, y_clf, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

regressors = {
    "Linear Regression": LinearRegression(),
}

for name, model in regressors.items():
    model.fit(X_train_reg, y_train_reg)
    print(f"{name} model trained.")
```

```python
from sklearn.linear_model import LogisticRegression
classifiers = {
    "Logistic Regression": LogisticRegression(),
}

for name, model in classifiers.items():
    model.fit(X_train_clf, y_train_clf)
    print(f"{name} model trained.")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

for name, model in classifiers.items():
    y_pred = model.predict(X_test_clf)
    print(f"{name} Evaluation:")
    print(f"Accuracy: {accuracy_score(y_test_clf, y_pred)}")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

for name, model in classifiers.items():
    y_pred = model.predict(X_test_clf)
    print(f"{name} Evaluation:")
    print(f"Precision: {precision_score(y_test_clf, y_pred)}")
```
